# Phase 2 — Data Collection: Web Scraping
Scraping programming language job demand data from a real jobs listing site.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import csv
```

```python
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/datasets/Programming_Languages.html"

response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

table = soup.find("table")

# Extract all rows including header
all_rows = table.find_all("tr")

# First row = headers
headers = [td.text.strip() for td in all_rows[0].find_all(["th", "td"])]

# Remaining rows = data
rows = []
for row in all_rows[1:]:
    cols = [td.text.strip() for td in row.find_all("td")]
    if cols:
        rows.append(cols)

df = pd.DataFrame(rows, columns=headers)
print(df.shape)
df.head(10)
```

```python
df.to_csv("../data/scraped_languages.csv", index=False)
print("✅ Saved to data/scraped_languages.csv")
```
