# Phase 2 — Data Collection: API Data
Collecting data on popular technologies and skills using the GitHub Jobs API dataset provided by IBM.

```python
import requests
import pandas as pd
```

```python
# Using IBM SkillsNetwork hosted jobs API
api_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/module%201/Accessing%20Data%20Using%20APIs/jobs.json"

response = requests.get(api_url)
print("Status Code:", response.status_code)

data = response.json()
print("Total records:", len(data))
print("Sample record:\n", data[0])
```

```python
df_jobs = pd.DataFrame(data)
print(df_jobs.shape)
df_jobs.head()
```

```python
import re

technologies = [
    "JavaScript", "SQL", "HTML", "Python", "Java", "MySQL",
    "Linux", "CSS", "PHP", "R", "Spark", "Scala",
    "AWS", "Excel", "MongoDB", "Tableau", "Power BI", "Git", "C"
]

tech_counts = {}
for tech in technologies:
    # Use word boundary to avoid partial matches
    pattern = rf"\b{re.escape(tech)}\b"
    count = df_jobs["Key Skills"].str.contains(pattern, case=False, na=False).sum()
    tech_counts[tech] = count

df_tech = pd.DataFrame(list(tech_counts.items()), columns=["Technology", "Job_Count"])
df_tech = df_tech.sort_values("Job_Count", ascending=False).reset_index(drop=True)
print(df_tech)
```

```python
# Split all skills and count the most frequent ones across all job postings
from collections import Counter

all_skills = df_jobs["Key Skills"].dropna().str.split("|").explode().str.strip()
top_skills = Counter(all_skills).most_common(20)

df_top_skills = pd.DataFrame(top_skills, columns=["Skill", "Count"])
print(df_top_skills)
```

```python
df_jobs.to_csv("../data/api_jobs.csv", index=False)
df_tech.to_csv("../data/api_tech_demand.csv", index=False)
df_top_skills.to_csv("../data/api_top_skills.csv", index=False)
print("✅ Saved all 3 files to data/")
```
